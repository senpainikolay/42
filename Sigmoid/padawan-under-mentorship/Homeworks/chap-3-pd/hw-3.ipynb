```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Pokemon.csv') 
df = df.drop(['#'], axis = 1) # Dropping the index column
```

```python
df.describe() # displaying general statistical data 
# Here we can easily see the RANGE  of the sample.
```

```python
# False = Non-Legendary
df.groupby(['Legendary']).mean()
```

```python
#
# Initializing dfs for .mode() method  
#
df_legendary = df   
df_non_legendary = df   

#
# Parsing and  dropping row's indexes according to the df's requirements 
#

for x in range (df.shape[0]):
    if df['Legendary'][x] == False:
        df_legendary = df_legendary.drop([x])  
    else:
        df_non_legendary = df_non_legendary.drop([x])   
        
```

```python
# Dropping the Name Column since we do not need them
df_legendary = df_legendary.drop(['Name'], axis = 1) 
df_non_legendary = df_non_legendary.drop(['Name'], axis = 1)
```

```python
# LEGENDARY sample
#
# Applying the .mode() method on the sample. #value that appears most often
# 

for x in df_legendary.columns:
    print( f" \n {x} : { df_legendary[x].mode() }") 
   
```

```python
# NON-LEGENDARY sample
#
# Applying the .mode() method on the sample. #value that appears most often
# 
for x in df_non_legendary.columns:
    print( f" \n {x} : { df_non_legendary[x].mode() }") 
```

```python
# We can observe that .median() method is approx/almost similar to .mean() meaning that the data is not skewed,it is normalized. 
df.groupby(['Legendary']).median()
```

```python
 df.groupby(['Legendary']).std() # Standard Deviation. Better results on Legendary (!! except the Sp/Speed Defense) -->
 # --> meaning that some Legendary pokeman have a higher coeficient of variety in speed Defense --> should be careful choosing
# a pokemon to fight with. More attention should be paid to the Sp. Def. detail on a legendary pokemon.
```

```python
df.groupby(['Legendary']).var() # Variantion =  that is bassically  .std() ^ 2 by the formula.
```
