```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart.csv')  
```

```python
# Mapping categorical data  to binary classification
df['ExerciseAngina'] = df['ExerciseAngina'].map({ 'N': 0, 'Y':1})    
df['Sex'] = df['Sex'].map({ 'M': 0, 'F':1})  

# Dummies for categorical data
df = pd.get_dummies(df, columns=['ChestPainType','RestingECG','ST_Slope'])
```

```python
""" 
    Positioning the dependent column as the last column.
"""
# Getting a copy of series of target column
temp_target = df['HeartDisease']   

# Removing the target column at the present position
df.pop('HeartDisease')  

# Adding back  the column target as the last column
df.insert(df.shape[1], 'HeartDisease', temp_target)
```

```python
abs ( df.corr() )
```

# Correlation between features.

```python
# Parsing the correlation table and selecting the features with highest correlations between them.  
#  > 0.4 corr() 
for i in range (df.shape[1] - 1):
    for j in range(i):
        if abs ( df.corr().values[i][j] ) > 0.3:
            print(f'{df.columns[i + 1 ]} and {df.columns[j + 1 ]} : { abs (df.corr().values[i][j] ) } ')
```

#### 1. (MaxHR and Sex)  A woman's heart pumps faster than a man's, but a man's heart ejects more blood with each pump 
#### 2. ( MaxHR and ExerciseAngina ) Target heart rate angina. Heart rate control is the first and most important step in order to achieve symptomatic control in patients with stable angina. A target heart rate is 55–60 beats per minute (bpm) 
#### 3. (Oldpeak and Exercise Angina)  Oldpeak means Exercise relative to rest. 
#### 4.  (Chestpain and Angina) Angina is a type of chest pain caused by reduced blood flow to the heart.  
#### 5.   The ST segment shift relative to exercise-induced increments in heart rate, the ST/heart rate slope ( slope), has been proposed as a more accurate ECG criterion for diagnosing significant coronary artery disease (CAD) which can have sympthoms as angina, chestpain.

# Creating a new dataframe with most correlated features to the target column 
# corr() > 0.3

```python
#initializing df's index array with highest correlations to the target column
index_cor_arr = list()


for i in range (df.shape[1] - 1 ):
    # each time assigning   val with  column's correlated value to the HeartDisease's index 
    val =   df.corr().values[ df.shape[1] - 2 ][i]  
    val = abs(val) 
    #comparing the val and storing the column's indexes in a list
    if val > 0.3:
        index_cor_arr.append(i)  

#Initializing a dataframe with correlated values ( corr coef is > 0.2 )  
df2 = pd.DataFrame({  
    
    # The column's names are taken from df and they are being assigned the values from df too, all in a loop.  
    f'{df.columns[ index_cor_arr[i] + 1 ] }' :  df[f'{ df.columns[ index_cor_arr[i] + 1  ] }'] for i in range ( len(index_cor_arr)  ) 
    
            })
```

```python
abs ( df2.corr() ) 
```

### The features in the table represents some logical assumptions about heartDisease and indicators of the disease ( such as slopes).

# Building the models

```python
# functions that takes a dataset as input and trains a model on LR
def log_r_score(df):
    """" 
    X : independent vars 
    y : dependent vars   
    
    """ 
    X = df.iloc[:, :-1].values
    y = df.iloc[:, -1].values 

    # Splitting the data in Training set and Testing's
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0) 

    # Fitting the data in the Log_R model
    from sklearn.linear_model import LogisticRegression
    classifier = LogisticRegression(random_state = 0)
    classifier.fit(X_train, y_train)

    # Predicting 
    y_pred = classifier.predict(X_test)   
    
    # Printing the confusion matrix and accuracy of the model
    from sklearn.metrics import confusion_matrix, accuracy_score
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    print(accuracy_score(y_test, y_pred))
```

```python
# All the features 
log_r_score(df)
```

```python
# Highest correlated features
log_r_score(df2)
```

# Conclusion: 
### 1.  The 2nd model, on highest correlated sample,  is theoretically faster and smaller in size but less accurate. As we can observe, there is a precision/accuracy difference of  just 0.5.   
### 2. By the confusion matrix, both models got the same incorrect score that a person have   heart dissease. And the second one predicted incorrectly more situations where persons do not have heart dissease. 
### 3.  Well, this example demonstrates the dylema of a ML engineer:   a faster model with less accuracy OR a slower one with more precision ?  
#### In this case, for the sake of medicine, would totally go with the 1st model.
