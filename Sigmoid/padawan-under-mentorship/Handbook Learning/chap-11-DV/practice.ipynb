```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
plt.figure(dpi = 150)
```

```python
x = np.linspace(0,10)
```

```python
plt.subplot(2,1,1) 
plt.plot(x,np.sin(x)) 
plt.subplot(2,1,2) 
plt.plot(x,np.cos(x))
```

```python
fig, ax = plt.subplots(2, dpi = 150) 
ax[0].plot(x, np.sin(x))
ax[1].plot(x,np.cos(x)) 
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure(dpi = 150)
ax = plt.axes() 

plt.plot(x, np.sin(x), color = 'red', label = 'sin(x)')
plt.plot(x,np.cos(x), color = 'green', linestyle = 'dotted', label ='cos(x)')

plt.xlim([-2,12]) 
plt.ylim([-1.25, 1.25]) 

plt.title('The sine and the cosine function')
plt.xlabel('x')
plt.ylabel('y') 

plt.legend()
```

```python
rng = np.random.RandomState(42) 
x = rng.randn(100) 
y = rng.randn(100) 
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.figure(dpi = 150) 
plt.xlim([-3,3])
plt.ylim([-3,3])
plt.scatter(x,y, c = colors, s = sizes, alpha = 0.3,cmap ='viridis') 
plt.colorbar()
```

```python
number_of_points = 70
x = np.linspace(0,10,number_of_points) 
dy = 0.25 
np.random.seed(17) 
y = np.sin(x) + dy * np.random.randn(number_of_points)

plt.figure(dpi = 150) 
plt.errorbar(x,y,yerr = dy, fmt =' .k')
```

```python
plt.style.use('seaborn-white')

def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y*x) * np.cos(x)

x = np.linspace(0,5,50) 
y = np.linspace(0,5,40) 

X, Y =np.meshgrid(x,y) 
Z = f(X,Y)

plt.figure(dpi = 150)
plt.contour(X,Y,Z,20,cmap = 'RdGy')
plt.colorbar() 
```

```python
np.random.seed(0) 
x1 = np.random.normal(0,0.8,1000) 
x2 = np.random.normal(-2,1,1000) 
x3 = np.random.normal(4,1,1000) 

plt.figure(dpi = 150) 
plt.hist(x1, alpha = 0.3, bins =40) 
plt.hist(x2, alpha = 0.3, bins =40) 
plt.hist(x3, alpha = 0.3, bins =40)
```

```python
mean = [0,0] 
cov = [[1,1], [1,2]] 
x,y = np.random.multivariate_normal(mean,cov,10000).T

plt.figure(dpi = 150)
plt.hist2d(x,y, bins = 30, cmap = 'Blues') 
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
import seaborn as sns
df = pd.read_csv('heart.csv') 

with sns.axes_style('whitegrid'):
    plt.figure(dpi=150)
    sns.kdeplot(df.loc[df['target'] == 0,'age'],
    shade = True, label = 'No Heart Disease') 
    sns.kdeplot(df.loc[df['target'] == 1, 'age'],
    shade = True, label = 'Heart Disease') 
    plt.legend()
```

```python
with sns.axes_style('darkgrid'):
    plt.figure(dpi=150)
    sns.kdeplot(data=df,x='age', y = 'trestbps') 
```

```python
with sns.axes_style('darkgrid'):
    sns.jointplot(data = df, x='age', y ='trestbps', kind = 'kde', height =7)

with sns.axes_style('white'):
    sns.jointplot(data=df,x='age',y='trestbps',kind='hex',height=7)
```

```python
plt.figure(dpi=150)
sns.boxplot(x='target', y='age',data=df)
```

```python
plt.figure(dpi= 150) 
sns.violinplot(x='target',y='age',data=df)
```

```python
sns.lmplot(data=df,x='age',y='chol',hue='target',
          height =7,aspect =1)
```
