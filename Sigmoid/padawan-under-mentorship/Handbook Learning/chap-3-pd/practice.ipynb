```python
import pandas as pd
import numpy as np 
```

```python
list_data = [np.random.random() for _ in range (200)]
```

```python
series = pd.Series(list_data)
print(series)
```

```python
for val in series:
    print(val)
```

```python
print(series.min()) 
print( series.max() )
```

```python
print(series.mean())  ################  the sum of quantities  divided by quantaty
```

```python
print(series.median())  ####################  sorts the values than takes the middle number if the len(series) is odd, else
                                         ##### if the len(series) is even than a avg(middle-1,middle+1) is the median value.
```

```python
print(series.mode()[:5]) ########## the most frequent values
```

```python
print(series.argmin()) # indexes
print(series.argmax())
```

```python
df =  pd.read_csv('heart.csv')
```

```python
df.describe() ## Categorical Feature = bounded range   / Numerical Feature = unbounded range.
```

```python
df[['sex','exang']].corr() ####  abs([:,:]) ->  1 and near 1 = similar   , 0,near 0 at all not
```

```python
for key in dict(df['sex'].value_counts()):
    print(f"{key} :  { df[ df['sex'] == key] ['chol'].mean() }" )  # groups by 'sex' the values in 'chol' and .mean() 
                                                                                                # is applied on 'chol' 
```

```python
df.groupby(['sex']).mean()
```

```python
df.groupby(['sex'])['chol'].mean()
```
